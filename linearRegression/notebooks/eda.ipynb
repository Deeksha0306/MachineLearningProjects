```python
# Importing all the necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading the dataset.
df = pd.read_csv("../data/raw_house_data.csv")
```

```python
# To search if any null value is there in dataset.
df.isnull().sum()
```

```python
df.columns
```

```python
# Dropping unecessary column.
df.drop('id', axis=1, inplace=True)
```

```python
# Converting date column values to date format.
df['date'] = pd.to_datetime(df['date'])
```

```python
# Checking the correlation among features and label.
correlation_matrix = df.corr()
```

```python
# Ploting a correlation heatmap.
plt.figure(figsize=(15,12))
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm')
plt.title("Correlation heatmap.")
plt.show()
```
