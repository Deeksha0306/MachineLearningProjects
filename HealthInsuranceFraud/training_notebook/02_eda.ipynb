```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../notebook/provider_fraud_features.csv")
```

```python
df.head()
```

```python
# fraud vs non fraud countplot
sns.countplot(x = 'PotentialFraud', data = df)
plt.title("Fraud vs non fraud")
plt.show()
```

```python
# boxplot to show "Do fraud providers perform more inpatient procedures on average than non-fraud providers?    "
sns.boxplot(
    x = 'PotentialFraud',
    y = 'In_AvgProcedureCount',
    data = df
)
plt.title("Inpatient avg patient vs fraud")
plt.show()
```

```python
# top 3 insights from this graph
# 1. Fraud providers have a higher median inpatient procedure count
# 2. Fraud providers show more consistent behavior
# 3. High procedure count alone does not imply fraud    
```

```python
# boxplot to show
sns.boxplot(
    x = 'PotentialFraud',
    y = 'In_PctFullDiagnosis',
    data = df
)
plt.title("Pct Full Diagnosis Slots (Inpatient) vs fraud")
plt.show()
```

```python
# top 3 insights that can be derived from the above graph
# 1. Fraud providers have a higher tendency to fully populate diagnosis slots in inpatient claims.
# 2. Diagnosis saturation is rare among non-fraud providers and usually appears as isolated outliers.
# 3. Consistent diagnosis over-documentation, rather than occasional extremes, is indicative of fraud.
```
