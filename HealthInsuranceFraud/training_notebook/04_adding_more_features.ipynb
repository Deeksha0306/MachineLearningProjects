```python
df_beneficiary['Age'] = 2010 - pd.to_datetime(df_beneficiary['DOB']).dt.year
```

```python
# Identify chronic condition columns
chronic_cols = [
    col for col in df_beneficiary.columns
    if 'ChronicCond' in col or col == 'RenalDiseaseIndicator'
]

# Convert them to numeric (force non-numeric to NaN, then fill with 0)
df_beneficiary[chronic_cols] = df_beneficiary[chronic_cols].apply(
    pd.to_numeric, errors='coerce'
).fillna(0)

# Now safely compute ChronicCount
df_beneficiary['ChronicCount'] = df_beneficiary[chronic_cols].sum(axis=1)
```

```python
df_beneficiary['TotalAnnualReimbursed'] = (df_beneficiary['IPAnnualReimbursementAmt'].fillna(0) + 
                                         df_beneficiary['OPAnnualReimbursementAmt'].fillna(0))

df_beneficiary['TotalAnnualDeductible'] = (df_beneficiary['IPAnnualDeductibleAmt'].fillna(0) + 
                                          df_beneficiary['OPAnnualDeductibleAmt'].fillna(0))
```

```python
in_claims = df_inpatient[['Provider', 'BeneID']]
out_claims = df_inpatient[['Provider', 'BeneID']]

# Combine inpatient + outpatient claims
claims_df = pd.concat([in_claims, out_claims], axis=0)

# Remove duplicates
claims_df = claims_df.drop_duplicates().reset_index(drop=True)

print(claims_df.head())
print("Shape:", claims_df.shape)
```

```python
provider_bene_map = claims_df[['Provider', 'BeneID']].drop_duplicates()

# Step 3: इस मैप को बेनिफिशियरी डेटा के साथ जोड़ें
provider_bene = provider_bene_map.merge(df_beneficiary, on='BeneID', how='left')
```

```python
# STEP 1: make sure df_beneficiary has derived columns
df_beneficiary['DOB'] = pd.to_datetime(df_beneficiary['DOB'], errors='coerce')
df_beneficiary['Age'] = 2010 - df_beneficiary['DOB'].dt.year

chronic_cols = [
    col for col in df_beneficiary.columns
    if 'ChronicCond' in col or col == 'RenalDiseaseIndicator'
]

df_beneficiary[chronic_cols] = df_beneficiary[chronic_cols].apply(
    pd.to_numeric, errors='coerce'
).fillna(0)

df_beneficiary['ChronicCount'] = df_beneficiary[chronic_cols].sum(axis=1)

df_beneficiary['TotalAnnualReimbursed'] = (
    df_beneficiary['IPAnnualReimbursementAmt'].fillna(0) +
    df_beneficiary['OPAnnualReimbursementAmt'].fillna(0)
)

df_beneficiary['TotalAnnualDeductible'] = (
    df_beneficiary['IPAnnualDeductibleAmt'].fillna(0) +
    df_beneficiary['OPAnnualDeductibleAmt'].fillna(0)
)

# STEP 2: provider–beneficiary mapping (claims_df MUST exist)
provider_bene_map = claims_df[['Provider', 'BeneID']].drop_duplicates()

# STEP 3: merge beneficiary features INTO provider_bene
provider_bene = provider_bene_map.merge(
    df_beneficiary[
        ['BeneID', 'Age', 'ChronicCount',
         'TotalAnnualReimbursed', 'TotalAnnualDeductible']
    ],
    on='BeneID',
    how='left'
)

# STEP 4: now aggregation will work
bene_provider_agg = provider_bene.groupby('Provider').agg(
    Bene_UniqueCount=('BeneID', 'nunique'),
    Bene_AvgAge=('Age', 'mean'),
    Bene_AvgChronicCount=('ChronicCount', 'mean'),
    Bene_PctChronic=('ChronicCount', lambda x: (x > 0).mean()),
    Bene_AvgReimbursed=('TotalAnnualReimbursed', 'mean'),
    Bene_AvgDeductible=('TotalAnnualDeductible', 'mean')
).reset_index()
```

```python
bene_provider_agg = provider_bene.groupby('Provider').agg(
    Bene_UniqueCount=('BeneID', 'nunique'),
    Bene_AvgAge=('Age', 'mean'),
    Bene_AvgChronicCount=('ChronicCount', 'mean'),
    Bene_PctChronic=('ChronicCount', lambda x: (x > 0).mean()),
    Bene_AvgReimbursed=('TotalAnnualReimbursed', 'mean'),
    Bene_AvgDeductible=('TotalAnnualDeductible', 'mean')
).reset_index()

bene_provider_agg.head()
```

```python
final_df = df.merge(
    bene_provider_agg,
    on='Provider',
    how='left'
)

final_df.fillna(0, inplace=True)
```
