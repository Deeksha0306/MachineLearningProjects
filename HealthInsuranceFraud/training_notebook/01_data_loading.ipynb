```python
df_inpatient['has_attending_physician'] = df_inpatient['AttendingPhysician'].notna().astype(int)
df_inpatient['has_operating_physician'] = df_inpatient['OperatingPhysician'].notna().astype(int)
df_inpatient['has_other_physician'] = df_inpatient['OtherPhysician'].notna().astype(int)
```

```python
df_inpatient['DeductibleAmtPaid'].head()
```

```python
df_inpatient['DeductibleAmtPaid'].nunique()
```

```python
df_inpatient['DeductibleAmtPaid'] = df_inpatient['DeductibleAmtPaid'].fillna(0)
```

```python
df_inpatient.isnull().sum()
```

```python
# making list of all the diagnosis codes columns
diagnosis_cols = [col for col in df_inpatient.columns if col.startswith("ClmDiagnosisCode")]
diagnosis_cols
```

```python
df_inpatient[diagnosis_cols].isnull().sum()
```

```python
df_inpatient['TotalDiagnosisCount'] = df_inpatient[diagnosis_cols].notnull().sum(axis=1)
df_inpatient['TotalDiagnosisCount']
```

```python
row_index = 0  # For the first row (indexing is zero-based)
claim_details = df_inpatient.iloc[row_index]
print(claim_details)
```

```python
procedure_cols = [col for col in df_inpatient.columns if col.startswith('ClmProcedureCode')]
total_procedure_codes = len(procedure_cols)
print(total_procedure_codes)
```

```python
df_inpatient['TotalProcedureCount'] = df_inpatient[procedure_cols].notnull().sum(axis=1)
df_inpatient['TotalProcedureCount']
```

```python
df_inpatient.isnull().sum()
```

```python
df_inpatient['ProcedureDiagnosisRatio'] = df_inpatient['TotalProcedureCount'] / (df_inpatient['TotalDiagnosisCount']+1)
df_inpatient['ProcedureDiagnosisRatio']
```

```python
df_inpatient['TotalProcedureCount'].describe()
```

```python
df_inpatient['TotalDiagnosisCount'].describe()
```

```python
df_inpatient['TotalProcedureCount'].quantile([0.95, 0.97, 0.99])  #95% of TotalProcedureCount is less than 2 and 5% is more than or equal to 2.
```

```python
df_inpatient['TotalDiagnosisCount'].quantile([0.7,0.90,0.91,0.99])
```

```python
max_diag = df_inpatient['TotalDiagnosisCount'].max()
df_inpatient['DiagnosisSlotsFull'] = (df_inpatient['TotalDiagnosisCount'] == max_diag).astype(int)
df_inpatient['DiagnosisSlotsFull'].sum()
```

```python
df_inpatient['HighProcedure_95'] = (df_inpatient['TotalProcedureCount'] > 2).astype(int)
```

```python
df_inpatient.columns
```

```python
provider_features = df_inpatient.groupby('Provider').agg(
    TotalClaims=('Provider', 'count'),
    AvgDiagnosisCount=('TotalDiagnosisCount', 'mean'),
    AvgProcedureCount=('TotalProcedureCount', 'mean'),
    PctFullDiagnosis=('DiagnosisSlotsFull', 'mean'),
    PctHighProcedure=('HighProcedure_95', 'mean'),
    AvgProcDiagRatio=('ProcedureDiagnosisRatio', 'mean')
).reset_index()
```

```python
provider_features.head()
```

```python
inpatient_provider = provider_features.rename(columns={
    'TotalClaims': 'In_TotalClaims',
    'AvgDiagnosisCount': 'In_AvgDiagnosisCount',
    'AvgProcedureCount': 'In_AvgProcedureCount',
    'PctFullDiagnosis': 'In_PctFullDiagnosis',
    'PctHighProcedure': 'In_PctHighProcedure',
    'AvgProcDiagRatio': 'In_AvgProcDiagRatio'
})
```

# Outpatient data

```python
df_outpatient = pd.read_csv(r"..\data\Train_Outpatientdata-1542865627584.csv")
```

```python
df_outpatient.head()
```

```python
df_outpatient.shape
```

```python
df_outpatient.columns
```

```python
df_outpatient.isnull().sum()
```

```python
df_outpatient['AttendingPhysician'].nunique()
```

```python
df_outpatient['has_attending_physician'] = df_outpatient['AttendingPhysician'].notna().astype(int)
df_outpatient['has_operating_physician'] = df_outpatient['OperatingPhysician'].notna().astype(int)
df_outpatient['has_other_physician'] = df_outpatient['OtherPhysician'].notna().astype(int)
```

```python
df_outpatient['DeductibleAmtPaid'] = df_outpatient['DeductibleAmtPaid'].fillna(0)
```

```python
diagnosis_cols = [col for col in df_outpatient.columns if col.startswith("ClmDiagnosisCode")]
diagnosis_cols
```

```python
df_outpatient['TotalDiagnosisCount'] = df_inpatient[diagnosis_cols].notnull().sum(axis=1)
df_outpatient['TotalDiagnosisCount']
```

```python
procedure_cols = [col for col in df_outpatient.columns if col.startswith('ClmProcedureCode')]
total_procedure_codes = len(procedure_cols)
print(total_procedure_codes)
```

```python
df_outpatient['TotalProcedureCount'] = df_outpatient[procedure_cols].notnull().sum(axis=1)
df_outpatient['TotalProcedureCount']
```

```python
df_outpatient['TotalProcedureCount'].unique()
```

```python
df_outpatient['TotalDiagnosisCount'].quantile([0.7,0.90,0.91,0.99])
```

```python
df_outpatient['TotalProcedureCount'].quantile([0.95, 0.97, 0.99])  #95% of TotalProcedureCount is less than 2 and 5% is more than or equal to 2.
```

```python
df_outpatient['HasProcedure'] = (df_outpatient['TotalProcedureCount'] > 0).astype(int)
```

```python
df_outpatient['Out_ProcedureDiagnosisRatio'] = df_outpatient['TotalProcedureCount'] / (df_outpatient['TotalDiagnosisCount']+1)
df_outpatient['Out_ProcedureDiagnosisRatio']
```

```python
max_diag = df_outpatient['TotalDiagnosisCount'].max()
df_outpatient['DiagnosisSlotsFull'] = (df_outpatient['TotalDiagnosisCount'] == max_diag).astype(int)
df_outpatient['DiagnosisSlotsFull'].sum()
```

```python
outpatient_provider = df_outpatient.groupby('Provider').agg(
    TotalClaims=('Provider', 'count'),
    AvgDiagnosisCount=('TotalDiagnosisCount', 'mean'),
    AvgProcedureCount=('TotalProcedureCount', 'mean'),
    PctFullDiagnosis=('DiagnosisSlotsFull', 'mean'),
    PctHasProcedure=('HasProcedure', 'mean'),
    AvgProcDiagRatio=('Out_ProcedureDiagnosisRatio', 'mean')
).reset_index()
```

```python
outpatient_provider = outpatient_provider.rename(columns={
    'TotalClaims': 'Out_TotalClaims',
    'AvgDiagnosisCount': 'Out_AvgDiagnosisCount',
    'AvgProcedureCount': 'Out_AvgProcedureCount',
    'PctFullDiagnosis': 'Out_PctFullDiagnosis',
    'PctHasProcedure': 'Out_PctHasProcedure',
    'AvgProcDiagRatio': 'Out_AvgProcDiagRatio'
})
```

```python
outpatient_provider.head()
```

```python
outpatient_provider.describe()
```

```python
provider_final = inpatient_provider.merge(
    outpatient_provider,
    on='Provider',
    how='outer'
)
```

```python
provider_final = provider_final.fillna(0)
```

```python
provider_final.shape
```

```python
provider_final.head()
```

```python
provider_final.describe()
```

```python
provider_final = provider_final.merge(
    df_provider[['Provider', 'PotentialFraud']],
    on = 'Provider',
    how = 'left'
)
```

```python
# covert yes to 1 and no to 0 in the provider table
provider_final['PotentialFraud'] = provider_final['PotentialFraud'].map({'Yes': 1, 'No': 0})
```

```python
provider_final.to_csv("provider_fraud_features.csv", index=False)
```
