```python
orders_data = {
    "date": pd.date_range(start="2025-01-01", periods=5, freq="D"),
    "product_id": [101, 102, 103, 104, 105],
    "city_id": [1, 2, 1, 3, 2],
    "orders": [10, 25, 7, 14, 30]
}

orders_df = pd.DataFrame(orders_data)
orders_df.head()
```

```python
import random
from datetime import datetime, timedelta
random_messages = [
    "Order received",
    "Processing order",
    "Order shipped",
    "Delivered successfully",
    "Delayed due to weather"
]

messages_data = {
    "date": [
        datetime.now() - timedelta(hours=i)
        for i in range(5)
    ],
    "msg": random.sample(random_messages, 5)
}

messages_df = pd.DataFrame(messages_data)
messages_df.head()
```

```python
messages_df["date"] = messages_df["date"].dt.floor("S")
messages_df.head()
```

```python
orders_df.info()
```

```python
orders_df['date_year'] = orders_df['date'].dt.year
orders_df['date_month'] = orders_df['date'].dt.month
orders_df['date_month_name'] = orders_df['date'].dt.month_name()
orders_df['date_day'] = orders_df['date'].dt.day
orders_df['date_day_name'] = orders_df['date'].dt.day_name()
orders_df['date_week'] = orders_df['date'].dt.isocalendar().week
orders_df['quarter'] = orders_df['date'].dt.quarter
orders_df['semester'] = np.where(orders_df['quarter']. isin([1,2]), 1,2)

orders_df.head()
```

```python
orders_df['isWeekend'] = np.where(orders_df['date_day_name'].isin(['Saturday','Sunday']), 1,0)
orders_df.drop(columns = ['product_id', 'city_id', 'orders'])
orders_df.head()
```

```python
# extract time elapsed between dates
today = datetime.today()
today
```

```python
(today - orders_df['date']).dt.days
```

```python
# Months passed
np.round((today - orders_df['date']) / np.timedelta64(1, 'D') / 30, 0)
```
